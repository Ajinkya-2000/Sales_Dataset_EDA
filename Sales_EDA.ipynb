```python
dfnew['Order Date'] = pd.to_datetime(dfnew['Order Date'])
```

```python
dfnew['Time'] = dfnew['Order Date'].dt.time
```

```python
dfnew
```

```python
dfnew['Time_range'] = 'hello'
```

```python
dfnew
```

```python
# Assigned value 'Till 6 am' for this condition
dfnew.loc[ dfnew['Time'] <= pd.to_datetime('06:00:00').time() , 'Time_range'] = 'Till 6am'
```

```python
# Assigned value '6am to 12 pm'
dfnew.loc[(dfnew['Time'] > pd.to_datetime('06:00:00').time()) & (dfnew['Time'] <= pd.to_datetime('12:00:00').time()) , 'Time_range'] = '6am to 12pm'
```

```python
dfnew.loc[(dfnew['Time'] > pd.to_datetime('12:00:00').time()) & (dfnew['Time'] <= pd.to_datetime('18:00:00').time()) , 'Time_range'] = '12pm to 6pm'
```

```python
dfnew.loc[(dfnew['Time'] > pd.to_datetime('18:00:00').time()) & (dfnew['Time'] <= pd.to_datetime('23:59:59').time()),'Time_range'] = '6pm to 12 am'
```
